```python
#importing pandas and numpy
import pandas as pd
import numpy as np
```

```python
#Reading the dataset
pth= r'C:\Users\Sujai\Desktop\Upgrad_docs\bike\day.csv'
bike = pd.read_csv(pth)
bike.head(10)
```

```python
#Changing the variable season as per season names
varlist=["season"]
def season_map(x):
    return x.map({'1':'spring','2':'summer', '3':'fall', '4':'winter'})

bike[varlist]=bike[varlist].astype(str).apply(season_map)
bike.head(20)
```

```python
#Changing the weather variable as per value
varlist=["weathersit"]
def weathersit_map(x):
    return x.map({'1':'clear','2':'mist+cloudy', '3':'Light Snow/Light Rain', '4':'Heavy Rain/Heavy Now'})

bike[varlist]=bike[varlist].astype(str).apply(weathersit_map)
bike.head(20)
```

```python
#dropping variable instant

bike=bike.drop(['instant'],axis=1)
```

```python
#importing data visualization modules
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
sns.pairplot(bike)
plt.show()
```

```python
plt.figure(figsize=(20, 12))
plt.subplot(2,4,1)
sns.boxplot(x = 'season', y = 'cnt', data = bike)
plt.subplot(2,4,2)
sns.boxplot(x = 'yr', y = 'cnt', data = bike)
plt.subplot(2,4,3)
sns.boxplot(x = 'mnth', y = 'cnt', data = bike)
plt.subplot(2,4,4)
sns.boxplot(x = 'holiday', y = 'cnt', data = bike)
plt.subplot(2,4,5)
sns.boxplot(x = 'weekday', y = 'cnt', data = bike)
plt.subplot(2,4,6)
sns.boxplot(x = 'workingday', y = 'cnt', data = bike)
plt.subplot(2,4,7)
sns.boxplot(x = 'weathersit', y = 'cnt', data = bike)
plt.show()
```

```python
#converting the string column season into dummies based on values
status = pd.get_dummies(bike['season'],drop_first=True)
status
```

```python
#converting the string column weathersit into dummies based on values

status_wether = pd.get_dummies(bike['weathersit'],drop_first=True)
status_wether
```

```python
#adding the newly created columns
bike = pd.concat([bike, status], axis = 1)
bike = pd.concat([bike, status_wether], axis = 1)
```

```python
bike.head(10)
```

```python
#Converting the true and false values to 0 and 1 , to perform analysis
def fall_map(x):
    return x.map({'True':0,'False':1})

varlist=["spring","summer","winter","clear","mist+cloudy"]

bike[varlist]=bike[varlist].astype(str).apply(fall_map)
bike.head(20)
```

```python
bike_1=bike.drop(["dteday","season","weathersit"],axis=1)
plt.figure(figsize = (16, 10))
sns.heatmap(bike_1.corr(), annot = True, cmap="YlGnBu")
plt.show()
```

```python
#Dropping all the unrequired columns
#Dropping the columns season and weathersit, since we have created new columns based on the values of these columns
#Dropping dteday, since there is no particular use for it for analysis, we already have mnth variable for month
#Dropping atemp because it won't be needed for analysis, we can use temp
#Casual+Registered=cnt, so this can cause multicolinearity, hence dropping both the variables
bike_1=bike.drop(['dteday','weathersit','season','atemp','casual','registered'],axis=1)
```

```python
from sklearn.model_selection import train_test_split

# We specify this so that the train and test data set always have the same rows, respectively
np.random.seed(0)
df_train, df_test = train_test_split(bike_1, train_size = 0.7, test_size = 0.3, random_state = 100)
```

```python
#Normalizing the variables temp, hum and windspeed, since it has high values
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
num_vars = ['temp', 'hum', 'windspeed']

df_train[num_vars] = scaler.fit_transform(df_train[num_vars])
df_train.head()
```

```python
#Plotting a heat map to check correlation
plt.figure(figsize = (16, 10))
sns.heatmap(df_train.corr(), annot = True, cmap="YlGnBu")
plt.show()
```

```python
#As we can see, the variable cnt seems to have high correlation with temp,yr,mnth and spring
```

```python
#Creating X-train and Y-train
y_train=df_train.pop("cnt")
X_train=df_train
```

```python
X_train.head()
```

```python
# Check for the VIF values of the feature variables. 
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif = pd.DataFrame()
vif['Features'] = X_train.columns
vif['VIF'] = [variance_inflation_factor(X_train.values, i) for i in range(X_train.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

```python
#Build a linear model

import statsmodels.api as sm
X_train_lm = sm.add_constant(X_train)

lr = sm.OLS(y_train, X_train_lm).fit()
lr.params
```

```python
lr.summary()
```

```python
#Since mnth has a high P value and a high vif, dropping the column mnth
X_train=X_train.drop('mnth',axis=1)
```

```python
X_train_lm_2 = sm.add_constant(X_train)

lr_2 = sm.OLS(y_train, X_train_lm_2).fit()
#lr.params
```

```python
lr_2.summary()
```

```python
#Training the model by predicting y values for X-train
y_train_price = lr_2.predict(X_train_lm_2)
```

```python
# Plot the histogram of the error terms
fig = plt.figure()
sns.distplot((y_train - y_train_price), bins = 20)
fig.suptitle('Error Terms', fontsize = 20)                  # Plot heading 
plt.xlabel('Errors', fontsize = 18)                         # X-labe
```

```python
#The error terma are normally distributed with mean at 0, which means that the model looks good
```
